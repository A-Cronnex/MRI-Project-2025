```python
import pandas as pd
```

```python
alzh_bp = pd.read_csv("clinical_data_id_age_gender.csv") #2 es mujer, 1 es hombre"
```

```python
df_prov = alzh_bp[(alzh_bp["Subject"] == 'tb')]
df_prov
```

```python
df_names = pd.Series(["tb_" + str(number) for number in df_prov["id"]], index = df_prov.index)
df_names
```

```python
df_prov["Subject"] = df_names
#df_prov.drop(labels ="id")
df_prov["diagnosis"] = "bipolar"
df_prov
```

```python
df_part = pd.read_csv("participants.tsv", sep="\t")[["participant_id","diagnosis","age","gender"]]
df_part
```

```python
df_part_n = df_part[(df_part["diagnosis"] == "BIPOLAR")]
```

```python
df_part_n["gender"] = df_part_n["gender"].map({"M":1,"F":2})
```

```python
df_part_n
```

```python
df_prov = df_prov.rename(columns = {"Subject":"participant_id", "Age":"age"})
df_prov
```

```python
df_COBRE = pd.read_csv("participants_COBRE.tsv", sep= "\t")
df_COBRE = df_COBRE.rename(columns = {"sex":"gender","dx":"diagnosis"})
df_COBRE["gender"] = df_COBRE["gender"].map({"male":1,"female":2})
```

```python
df_COBRE
```

```python
SRBPNS = pd.read_csv("participants_SRBPNS.tsv", sep = "\t")
```

```python
SRBPNS = SRBPNS[["participant_id","diag","age","sex"]]
```

```python
SRBPNS = SRBPNS.rename(columns = {"sex":"gender"})
SRBPNS = SRBPNS[(SRBPNS["diag"] == 7)]
```

```python
SRBPNS = SRBPNS.rename(columns = {"diag":"diagnosis"})
SRBPNS
```

```python
final_dataframes = [df_part_n,df_COBRE,df_prov,SRBPNS]
result = pd.concat(final_dataframes)
result = result.drop(columns=['study','id'])
result["diagnosis"] = 7 #trastorno bipolar equivale a 7
```

```python
import re

def clean_mrid(value):
    if value.startswith("tb_"):
        return value
    if not value.startswith("sub-"):
        value = f"sub-{value}"
    return re.match(r"(sub-[^_]+)", value).group(1)

# Apply to the MRID column
result['participant_id'] = result['participant_id'].apply(clean_mrid)
```

```python
result = result.rename(columns = {'participant_id': 'MRID', 'age':'Age','gender':'Sex'})
result
```

```python
result.to_csv("NAMES_BIPOLAR.csv", index = False)
```
