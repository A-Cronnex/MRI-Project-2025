```python
import pandas as pd
```

```python
first_csv = pd.read_csv("2025-05-04Bipolar_1.csv")
```

```python
second_csv = pd.read_csv("2025-05-04Bipolar_2.csv")
```

```python
second_csv
```

```python
final_dataframe = [first_csv, second_csv]
final = pd.concat(final_dataframe)
final = final.drop(columns = ["Unnamed: 0"])
```

```python
final.to_csv("BIPOLAR_ROIs.csv", index=False)
```

```python
import re

def clean_mrid(value):
    if value.startswith("tb_"):
        return value
    if not value.startswith("sub-"):
        value = f"sub-{value}"
    return re.match(r"(sub-[^_]+)", value).group(1)

# Apply to the MRID column
final['MRID'] = final['MRID'].apply(clean_mrid)
```
